### Exploring potential connections between age and family status in terms of ordering habits

```python
# Age groups making up each family status group

ct = pd.crosstab(ords_prods_all['fam_status'], ords_prods_all['age_group'])
age_fam = ct.plot(color = age_colors, kind='barh', stacked=True)
plt.ylabel("Family Status Groupings")
plt.title('Age Groups in Family Statuses')
```

```python
age_fam.figure.savefig(os.path.join(path, 'Analysis', 'Visualizations', 'age_fam_status.png'))
```

```python
# Exploring relationship between age_group, fam_status, and orders_day_of_the_week

pivot_table = pd.pivot_table(ords_prods_all, values='order_id', index=['age_group', 'fam_status'], columns='orders_day_of_the_week', aggfunc='count')
```

```python
plt.figure(figsize=(12, 8))
heatmap_age_fam_dow = sns.heatmap(pivot_table, cmap='YlGnBu', annot=True, fmt='d', cbar=True)
plt.title('Relationship between Age Group, Family Status, and Orders Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Age Group - Family Status')
plt.show()
```

```python
heatmap_age_fam_dow.figure.savefig(os.path.join(path, 'Analysis', 'Visualizations', 'heatmap_age_fam_dow.png'))
```

```python
# Exploring relationship between age_group, fam_status, and order_hour_of_day

pivot_table2 = pd.pivot_table(ords_prods_all, values='order_id', index=['age_group', 'fam_status'], columns='order_hour_of_day', aggfunc='count')
```

```python
plt.figure(figsize=(12, 8))
heatmap_age_fam_hod = sns.heatmap(pivot_table2, cmap='YlGnBu', annot=False, fmt='d', cbar=True)
plt.title('Relationship between Age Group, Family Status, and Order Hour of Day')
plt.xlabel('Order Hour of Day')
plt.ylabel('Age Group - Family Status')
plt.show()
```

```python
heatmap_age_fam_hod.figure.savefig(os.path.join(path, 'Analysis', 'Visualizations', 'heatmap_age_fam_hod.png'))
```
