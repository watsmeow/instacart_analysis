```python
orders_products_combined = pd.read_pickle(os.path.join(path, 'Data', 'Prepared Data', 'orders_products_combined.pkl'))
```

```python
df_prods = pd.read_csv(os.path.join(path, 'Data', 'Prepared Data', 'products_checked.csv'), index_col = False)
```

## 4. Check the shape of the imported dataframe

```python
# Deleting previous '_merge' column

ords_prods_combined = orders_products_combined.drop(columns = ['_merge'])
```

```python
# Deleting unnamed column in products dataframe

df_prods_clean = df_prods.drop(columns = ['Unnamed: 0'])
```

```python
df_prods_clean.head()
```

## 5. Determine a suitable way to combine the orders_products_combined dataframe with your products data set

```python
# Merge dataframes on 'product_id' column

orders_products_merged = ords_prods_combined.merge(df_prods_clean, on = 'product_id', how = 'left', indicator = True)
```

```python
orders_products_merged.shape
```

```python
orders_products_merged.head()
```
