## 3) Create a regional segmentation of the data & analyze regional spending habits

```python
# Delineating regions

Northeast = ['Maine', 'New Hampshire', 'Vermont', 'Massachusetts', 'Rhode Island', 'Connecticut', 'New York', 'Pennsylvania', 'New Jersey']
Midwest = ['Wisconsin', 'Michigan', 'Illinois', 'Indiana', 'Ohio', 'North Dakota', 'South Dakota', 'Nebraska', 'Kansas', 'Minnesota', 'Iowa', 'Missouri']
South = ['Delaware', 'Maryland', 'District of Columbia', 'Virginia', 'West Virginia', 'North Carolina', 'South Carolina', 'Georgia', 'Florida', 'Kentucky', 'Tennessee', 'Mississippi', 'Alabama', 'Oklahoma', 'Texas', 'Arkansas', 'Louisiana']
West = ['Idaho', 'Montana', 'Wyoming', 'Nevada', 'Utah', 'Colorado', 'Arizona', 'New Mexico', 'Alaska', 'Washington', 'Oregon', 'California', 'Hawaii']
```

```python
# Assign States in 'state' column to a region

cust_ords_prods.loc[cust_ords_prods['state'].isin(Northeast), 'region'] = 'Northeast'
cust_ords_prods.loc[cust_ords_prods['state'].isin(Midwest), 'region'] = 'Midwest'
cust_ords_prods.loc[cust_ords_prods['state'].isin(South), 'region'] = 'South'
cust_ords_prods.loc[cust_ords_prods['state'].isin(West), 'region'] = 'West'
```

```python
# Check outcome

cust_ords_prods['region'].value_counts(dropna = False)
```

```python
cust_ords_prods.shape
```

```python
cust_ords_prods.columns
```

```python
# Check for variations in spending habits by region - will vizualize in later steps

crosstab = pd.crosstab(cust_ords_prods['region'], cust_ords_prods['spending_flag'], dropna = False)
```

## 4) Create an exclusion flag for low-activity customers

```python
# Create flag for customers with less than 5 orders

cust_ords_prods.loc[cust_ords_prods['max_order'] < 5, 'customer_activity'] = 'low activity'
```

```python
# Create flag for customers with 5 orders or more

cust_ords_prods.loc[cust_ords_prods['max_order'] >= 5, 'customer_activity'] = 'high activity'
```

```python
cust_ords_prods['customer_activity'].value_counts(dropna = False)
```

```python
# Assign low activity profiles to new dataframe

low_activity_customers = cust_ords_prods[cust_ords_prods['customer_activity'] == 'low activity']
```

```python
low_activity_customers.head()
```

```python
# Export and save low activity profiles

low_activity_customers.to_pickle(os.path.join(path, 'Data', 'Prepared Data', 'low_activity_customers.pkl'))
```

```python
# Assign high activity profiles to new dataframe

high_activity_customers = cust_ords_prods[cust_ords_prods['customer_activity'] == 'high activity']
```

```python
high_activity_customers.shape
```

```python
# Export and save high activity profiles

high_activity_customers.to_pickle(os.path.join(path, 'Data', 'Prepared Data', 'high_activity_customers.pkl'))
```
