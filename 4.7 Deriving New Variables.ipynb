### If-Statements with User-Defined Functions

```python
# Create a function to assign price range to products

def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Apply 'price_label' function

df['price_range'] = df.apply(price_label, axis=1)
```

```python
# Check results of function application

df['price_range'].value_counts(dropna = False)
```

```python
# Check max price in subset

df['prices'].max()
```

# 2) Create a new column for 'busiest days' using a suitable method

```python
# Assign a string value to order days that will label order days according to business, delineating by the two most busiest days and two least busiest days

busiest_days_result = []

for value in ords_prods_merged["orders_day_of_the_week"]:
  if value == 0 or value == 1:
    busiest_days_result.append("Busiest day")
  elif value == 4 or value == 3:
    busiest_days_result.append("Least busy")
  else:
    busiest_days_result.append("Regularly busy")
```

```python
ords_prods_merged['busiest_days'] = busiest_days_result
```

# 4) Create a new column containing “Most orders,” “Average orders,” and “Fewest orders" labels called “busiest_period_of_day"

```python
# Check for busiest times of day

ords_prods_merged['order_hour_of_day'].value_counts(dropna = False)
```

```python
# Create arrays for holding results and for hour business to use in for loop

order_hour_info = []
busiest_hours = [10, 11, 14, 15, 13, 12, 16, 9]
slowest_hours = [23, 6, 0, 1, 5, 2, 4, 3]
```

```python
# Assign a string value to order hours that will label order hours according to “Most orders,” “Average orders,” and “Fewest orders"


for value in ords_prods_merged["order_hour_of_day"]:
  if value in busiest_hours:
    order_hour_info.append("Most orders")
  elif value in slowest_hours:
    order_hour_info.append("Fewest orders")
  else:
    order_hour_info.append("Average orders")
```

```python
order_hour_info
```

```python
ords_prods_merged['busiest_period_of_day'] = order_hour_info
```
